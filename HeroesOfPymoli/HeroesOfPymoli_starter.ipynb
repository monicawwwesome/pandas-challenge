```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

## Player Count

* Display the total number of players


```python
uni_player = purchase_data["SN"].unique()
num_player = len(uni_player)
num_player_df = pd.DataFrame({"Totla Number of Player":[num_player]})
num_player_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
num_item = purchase_data["Item ID"].nunique()
avg_price = purchase_data["Price"].mean()
num_purchases = purchase_data["Purchase ID"].nunique()
tot_revenue = purchase_data["Price"].sum()
pur_anal = {"Number of Unique Items":[num_item],
                     "Average Price":[avg_price],
                     "Number of Purchases":[num_purchases],
                     "Total Revenue":[tot_revenue]}
pur_anal_df = pd.DataFrame(pur_anal)
pur_anal_df["Average Price"] = pur_anal_df["Average Price"].map("${:.2f}".format)
pur_anal_df["Total Revenue"] = pur_anal_df["Total Revenue"].map("${:.2f}".format)
pur_anal_df.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_data = purchase_data.groupby("Gender")
num_gender = gender_data.nunique()['SN']
pct_gender = num_gender/num_player*100

gender_player = {"Total Count":(num_gender),
                "Percentage of Players":(pct_gender)}
gender_player_df = pd.DataFrame(gender_player)
gender_player_df["Percentage of Players"] = gender_player_df["Percentage of Players"].map("{:.2f}%".format)
gender_player_df
```

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
num_purchase = gender_data["Purchase ID"].count()
avg_price = gender_data["Price"].mean()
total_price = gender_data["Price"].sum()
avg_price_pp = total_price/num_gender
gender_anal = {"Purchase Count":(num_purchase),
               "Average Purchase Price":(avg_price),
               "Total Purchase Value":(total_price),
               "Avg Total Purchase per Person":(avg_price_pp)}
gender_anal_df = pd.DataFrame(gender_anal)
gender_anal_df["Average Purchase Price"] = gender_anal_df["Average Purchase Price"].map("${:.2f}".format)
gender_anal_df["Total Purchase Value"] = gender_anal_df["Total Purchase Value"].map("${:.2f}".format)
gender_anal_df["Avg Total Purchase per Person"] = gender_anal_df["Avg Total Purchase per Person"].map("${:.2f}".format)
gender_anal_df.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9,14,19,24,29,34,39,100]
age_group = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
purchase_data["Age Group"]= pd.cut(purchase_data["Age"],bins,labels=age_group)
age_data = purchase_data.groupby("Age Group")
num_age = age_data.nunique()["SN"]
pct_age = num_age/num_player*100

age_player = {"Total Count":(num_age),
                "Percentage of Players":(pct_age)}
age_player_df = pd.DataFrame(age_player)
age_player_df["Percentage of Players"] = age_player_df["Percentage of Players"].map("{:.2f}%".format)
age_player_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
num_purchase_age = age_data["Purchase ID"].count()
avg_price_age = age_data["Price"].mean()
total_price_age = age_data["Price"].sum()
avg_price_pp_age = total_price_age/num_age
age_anal = {"Purchase Count":(num_purchase_age),
               "Average Purchase Price":(avg_price_age),
               "Total Purchase Value":(total_price_age),
               "Avg Total Purchase per Person":(avg_price_pp_age)}
age_anal_df = pd.DataFrame(age_anal)
age_anal_df["Average Purchase Price"] = age_anal_df["Average Purchase Price"].map("${:.2f}".format)
age_anal_df["Total Purchase Value"] = age_anal_df["Total Purchase Value"].map("${:.2f}".format)
age_anal_df["Avg Total Purchase per Person"] = age_anal_df["Avg Total Purchase per Person"].map("${:.2f}".format)
age_anal_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
SN_data = purchase_data.groupby("SN")
num_pur_SN = SN_data["Item ID"].count()
avg_price_SN = SN_data["Price"].mean()
total_price_SN = SN_data["Price"].sum()
top_spender = {"Purchase Count":(num_pur_SN),
               "Average Purchase Price":(avg_price_SN),
               "Total Purchase Value":(total_price_SN)}
top_spender_df = pd.DataFrame(top_spender)
top_spender_df["Total Purchase Value"] = top_spender_df["Total Purchase Value"].astype(float)
top_spender_df["Total Purchase Value"] = top_spender_df["Total Purchase Value"].map("${:,.2f}".format)
top_spender_df["Average Purchase Price"] = top_spender_df["Average Purchase Price"].map("${:,.2f}".format)
top_spender_df.sort_values('Total Purchase Value', ascending = False).head(5)

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_data = purchase_data.groupby("Item ID")
item_name = item_data["Item Name"].unique()
num_pur_item = item_data["Item ID"].count()
item_price = item_data["Price"].mean()
total_price_item = item_data["Price"].sum()
pop_item = {"Item Name":(item_name),
            "Purchase Count":(num_pur_item),
            "Item Price":(item_price),
            "Total Purchase Value":(total_price_item)}
pop_item_df = pd.DataFrame(pop_item)
pop_item_df["Item Price"] = pop_item_df["Item Price"].map("${:.2f}".format)
pop_item_df["Total Purchase Value"] = pop_item_df["Total Purchase Value"].map("${:.2f}".format)
pop_item_df.sort_values("Purchase Count", ascending = False).head(5)
```
